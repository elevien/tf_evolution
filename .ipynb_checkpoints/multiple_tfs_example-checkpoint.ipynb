```python
import random,copy,math,time,os,sys

from importlib import reload

#import pandas as pd
import thermodynamic_model as tfe # this is the module I wrote 

# numpy/scipy stuff
import numpy as np

# plotting stuff
import matplotlib.pyplot as plt
```

# Make model with 2 TFs and 2 Binding sites

```python
L = 14 # number of base pairs per site
# generate energy matrix for each TF
# units are in kbT
E1 = np.random.normal(0.15,0.1,(L,4)) 
E1[:,1] = 0.1*np.ones(L)

E2 = np.random.normal(0.15,0.1,(L,4)) 
E2[:,0] = 0.1*np.ones(L)

fig,axs = plt.subplots(ncols =2,figsize=(10,5),sharex=True,sharey=True)
ax= axs[0]
ax.imshow(E1.T)
ax.set_xlabel(r'Position')
plt.yticks(np.arange(4), ('A','C','G','T'));

ax= axs[1]
ax.imshow(E2.T)
ax.set_xlabel(r'Position')
plt.yticks(np.arange(4), ('A','C','G','T'));
```

```python
# setup model paramaters
M_sites = 2 # number of binding sites
M_factors = 2 # number of transcription factors
binding_energies = [E1,E2] # binding energy matrices for each tf (a list of M_factors L x 4 arrays)
interaction_energies = np.array([[0.0,2.0],[0.0,0.0]]) # TF1 interacts with TF2
rates = np.ones((M_sites,M_factors))  # transcription rates
tf_concentrations = np.array([1.,1.]) # tf concentrations
```

```python
# define out model
reload(tfe)
model = tfe.tf_thermodynamic_model(binding_energies,interaction_energies,rates,tf_concentrations,M_sites)
sel= 0.05 # strength of selection on gene expression at site
def fitness_func(model,g): # use expression level as proxy for fitness
    return sel*model.expression_level(g)
```

```python
# lets make a random geneome and test out fitness function
g0 = np.array([np.random.choice(range(4),L) for k in range(M_sites)])
fitness_func(model,g0)
```

# Make fitness landscape

```python
g1 = np.random.choice(range(4),L)
g_set = [np.array([g1,np.random.choice(range(4),L)]) for k in range(1000)]
f_set = [fitness_func(model,g) for g in g_set]
E_set1 = [np.sum([binding_energies[0][k,g[1][k]] for k in range(L)]) + \
        np.sum([binding_energies[1][k,g[1][k]] for k in range(L)])  \
        for g in g_set]
```

```python
plt.plot(E_set1,f_set,'o',alpha=0.5)
ax = plt.gca()
ax.set_xlabel(r'$E$')
ax.set_ylabel('Fitness')
plt.legend(loc='upper right',frameon=False)
```

```python
# let's also try the case with zero interaction energy
model.interaction_energies  = np.zeros((2,2))
f_set = [fitness_func(model,g) for g in g_set]
E_set1 = [np.sum([binding_energies[0][k,g[1][k]] for k in range(L)]) + \
        np.sum([binding_energies[1][k,g[1][k]] for k in range(L)])  \
        for g in g_set]

plt.plot(E_set1,f_set,'o',alpha=0.5)
ax = plt.gca()
ax.set_xlabel(r'$E$')
ax.set_ylabel('Fitness')
plt.legend(loc='upper right',frameon=False)

# reset interaction energies 
interaction_energies = np.array([[0.0,2.0],[0.0,0.0]])
```

The fitness is no longer defined uniquely by the energy when there are multiple TFs and binding sites. This is because having, e.g. intermediate energy for both factors can be different than high energy for one and low of the other. Thus the total energy $E_1 + E_2$ does not uniquely define the fitness
